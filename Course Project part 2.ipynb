```python
import numpy as np
```

```python
# варіант №12   К-сть міст = 12    Діапазон місяців = 5-12     Регіон 1 = Полтава     Регіон 2 = Одеса
```

```python
# Одеса діапазон дзвінків = 100-250
```

```python
# Полтава діапазон дзвінків = 100 - 180
```

```python
# Перший регіон. 
# Кожна строчка даних по першому регіону відповідає місту регіону 
# (для парних варіантів). Кожен стовбець даних по першому регіону відповідає місяцю 2018 року (в
# діапазоні по варіанту, для парних варіантів)

CitiesCount = 12
MonthCount = 8
pMin = 100
pMax = 180
PoltavaRegion = np.random.randint(pMin, pMax, size = [CitiesCount, MonthCount], dtype = int)
PoltavaRegion
```

```python
# Другий регіон.
# Кожна строчка даних по першому регіону відповідає місяцю 2018 року (10
# місяців) для парних варіантів
# Кожен стовбець даних по другому регіону відповідає кількості міст регіону
# (по варіанту, для парних варіантів) + 20
MounthCount = 10
CitiesCount = CitiesCount + 20
oMin = 100
oMax = 250
OdesaRegion = np.random.randint(oMin, oMax, size = [MounthCount, CitiesCount], dtype = int)
OdesaRegion
```

```python
# 1. Згенерувати два ndarray (матриці) по заданим регіонам, відповідно заданої
# у описі предметної області структурі та відповідної розмірності. Комірки
# матриць зберігають тривалості дзвінків абонентів у роумінгу (згенерувати
# значення в зазначеному діапазоні згідно назвам регіонів в таблиці Б.2).

# 2. Створити новий (загальний) ndarray для зберігання всієї інформації з двох
# заданих регіонів за вказаний в таблиці Б.3 діапазон місяців, виконавши
# операції зрізу, трансформації (рядки нової матриці - міста, стовпці - місяці)
# і операції об'єднання.
# Місяці по варіанту = 6-9

AllRegions = np.vstack((PoltavaRegion[:,1:5],np.transpose(OdesaRegion)[:,5:9]))
AllRegions
```

```python
# На основі нового набору

# 3. Розрахувати сумарну, максимальну та мінімальну тривалості дзвінків для
# кожного місяця і середню кількість дзвінків по кожному місту.
np.sum(AllRegions,axis = 0)
```

```python
np.max(AllRegions,axis = 0)
```

```python
np.min(AllRegions,axis = 0)
```

```python
np.average(AllRegions, axis = 1)
```

```python
# 4. Розрахувати загальну (сумарну) тривалість дзвінків за три останніх місяці
# зазначеного в таблиці Б.3 діапазону по всьому набору даних.

np.sum(AllRegions[:,-3:], axis = 0)
```

```python
# 5. Припустимо, що ви отримали 10 окремих рядків, які зберігають інформацію по 10 новим містам 
# за всі місяці 2017 року і минулі місяці 2018 (січень - жовтень) - 22 значення в кожному отриманому рядку.

NewData = np.random.randint(pMin,oMax, size = [10,22] )
NewData
```

```python
#6. Додайте нову інформацію в набір, вибравши інформацію тільки за ті місяці,
#які вам необхідні (таблиця Б.3)

NewAllRegions = np.vstack((NewData[:,6:10],AllRegions))
NewAllRegions
```

```python
# 7. Використовуючи універсальні функції, вивести максимальну та мінімальну
# тривалості дзвінків по кожному місту і місяць (визначити назву міста та
# місяця за номером), коли вона була отримана.

np.argmax(NewAllRegions, axis = 1)
```

```python
np.max(NewAllRegions, axis = 1)
```

```python
np.argmin(NewAllRegions, axis = 1)
```

```python
np.min(NewAllRegions, axis = 1)
```

```python
# 8. За кожний місяць сформувати твердження: «В середньому в місяці Х
# тривалість дзвінків в роумінгу становила Y плюс \ мінус Z хвилин». Y і Z
# визначити за допомогою відповідних універсальних функцій. Х - номер
# місяця (від 1 до 12), а не номер стовпчика.
for i in range(0,4):
    print('В середньому, в місяці ', (i+6) ,' тривалість дзвінків у роумінгу становила ', np.average(NewAllRegions,axis=0)[i], ' плюс\мінус ', np.std(NewAllRegions[:,i]) )
```

```python
# 9. Для кожного міста та місяця визначити і вивести таке значення тривалості
# дзвінків L, що все значення тривалості дзвінків для даного міста (місяця)
# менше L з ймовірністю 60%.

np.percentile(NewAllRegions, 60, axis=0)
```

```python
np.percentile(NewAllRegions, 60, axis=1)
```
